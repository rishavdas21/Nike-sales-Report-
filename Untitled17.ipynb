```python
# Create a Visualization of Univariate Analysis of the Top 10 Brand Names
Products = df['Product_Name'].value_counts().reset_index().sort_values(by = 'count', ascending = False)[:10]
plt.figure(figsize=(8,6))
plt.bar(Products['Product_Name'], Products['count'], color='orange')
plt.xlabel('Product_Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.title('Top 10 Product in Market')
plt.legend(df['Product_Name'])
plt.show()
```

```python
# Average Selling MRP of the Products
Avg_MRP = df['MRP'].mean()
print(Avg_MRP)
```

```python
# MRP Distribution of the Products
plt.figure(figsize=(8,6))
sns.histplot(df['MRP'], bins = 50, kde=True)
plt.title("MRP Distribution of the Products")
plt.legend(df['MRP'])
plt.show()
```

```python
# Top 10 Products
PRODUCT = df['Product_Name'].value_counts()[:10].reset_index().sort_values(by='count', ascending=False)
print(PRODUCT)
```

```python
# Visualization of Top 10 Products "Pants and Trousers" Purchase by Consumers
plt.figure(figsize=(8,6))
plt.pie(PRODUCT['count'][:10], labels = PRODUCT['Product_Name'][:10], autopct = '%.0f%%')
plt.show()
```

```python
# Revenue vs Product Name Distribution of Nike Sales
plt.figure(figsize=(12, 6))
plt.scatter(x=df['Product_Name'], y=df['Revenue'])
plt.title("Revenue vs Product Name Distribution of Nike Sales")
plt.xlabel('Product Name')
plt.ylabel('Revenue')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
